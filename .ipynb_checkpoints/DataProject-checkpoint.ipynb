```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\adird\data.csv')
df
```

```python
df['Success'] = 0

# Compare 'Goal Amount' and 'Pledged Amount' and update 'New Column' accordingly
df.loc[df['Goal Amount'] > df['Pledged Amount'], 'Success'] = 1

# Save the updated DataFrame to a new CSV file
df.to_csv('updated_file.csv', index=False)
df
```

```python
import matplotlib.pyplot as plt


success_count = df[df['Success'] == 1]['Success'].count()
failure_count = df[df['Success'] == 0]['Success'].count()


# Calculate percentages
total = success_count + failure_count
success_percent = (success_count / total) * 100
failure_percent = (failure_count / total) * 100

# Create the pie chart
labels = ['Success', 'Failure']
sizes = [success_percent, failure_percent]
colors = ['#66BB6A', '#EF5350']  # Green and red colors
explode = (0.1, 0)  # Explode the 'Success' slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Success Percentage')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the pie chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame


# Sort the DataFrame by the "Percent Funded" parameter in descending order
df_sorted = df.sort_values(by='Percent Funded', ascending=False)

# Extract the percent funded values
percent_funded = df_sorted['Percent Funded']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create the bar graph
plt.bar(range(len(percent_funded)), percent_funded, color='green')

# Add labels and title
plt.xlabel('Project Index')
plt.ylabel('Percent Funded')
plt.title('Projects by Percent Funded')

# Show the bar graph
plt.show()
import pandas as pd


# Find the project with the highest percent funded value
max_percent_funded = df['Percent Funded'].max()
project_with_max_percent_funded = df[df['Percent Funded'] == max_percent_funded]

# Print the project name and percent funded
print(f"Project with the highest percent funded:")
print(project_with_max_percent_funded[['Project Name', 'Percent Funded']])


```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Count the occurrences of each currency parameter
usd_count = df[df['Currency'] == 'USD'].shape[0]
gbp_count = df[df['Currency'] == 'GBP'].shape[0]
eur_count = df[df['Currency'] == 'EUR'].shape[0]
else_count = df[~df['Currency'].isin(['USD', 'GBP', 'EUR'])].shape[0]

# Set the figure size
plt.figure(figsize=(8, 6))

# Define the labels and sizes for the pie chart
labels = ['USD', 'GBP', 'ELSE', 'EUR']
sizes = [usd_count, gbp_count, else_count, eur_count]
colors = ['#66BB6A', '#EF5350', '#FFB74D', 'blue']  # Green, Red, Orange

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Currency Distribution')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the pie chart
plt.show()
```
